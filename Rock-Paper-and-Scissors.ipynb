# Welcome to Rock, Paper and Scissors Game

```python
import numpy as np

print('Welcome to Rock Paper Scissor Game')
player_name = input('Enter your name : ')
print('Press')
print('r for rock\np for paper\ns for scissor')

game_list = ['r','p','s']
moves = 5
no_of_moves = 0
computer_score = 0
player_score = 0


while no_of_moves < moves:
    player_input = input('Choose one Rock,Paper,Scissor : ')
    player_input = player_input.lower()
    com_input = np.random.choice(game_list)
    
    if player_input==com_input:
        print('Tie Both got 0 points')
        
    elif player_input=='r' and com_input=='s':
        player_score = player_score +1
        print(f"your choice {player_input} and computer choice {com_input} \n")
        print(f"{player_name} won this round")
        print(f"{player_name} score is {player_score} and computer score is {computer_score}")
        
    elif player_input =='p' and com_input=='r':
        player_score = player_score +1
        print(f"your choice {player_input} and computer choice {com_input} \n")
        print(f"{player_name} won this round")
        print(f"{player_name} score is {player_score} and computer score is {computer_score}")
        
    elif player_input == 's' and com_input == 'p':
        player_score = player_score +1
        print(f"your choice {player_input} and computer choice {com_input} \n")
        print(f"{player_name} won this round")
        print(f"{player_name} score is {player_score} and computer score is {computer_score}")
        
    elif player_input == 'r' and com_input == 'p':
        computer_score = computer_score +1
        print(f"your choice {player_input} and computer choice {com_input} \n")
        print("computer won this round")
        print(f"{player_name} score is {player_score} and computer score is {computer_score}")    
    elif player_input == 'p' and com_input == 's':
        computer_score = computer_score +1
        print(f"your choice {player_input} and computer choice {com_input} \n")
        print("computer won this round")
        print(f"{player_name} score is {player_score} and computer score is {computer_score}") 
        
    elif player_input == 's' and com_input == 'r':
        computer_score = computer_score +1
        print(f"your choice {player_input} and computer choice {com_input} \n")
        print("computer won this round")
        print(f"{player_name} score is {player_score} and computer score is {computer_score}") 
        
    else:
         print("Wrong choice Please try again\n")
    no_of_moves = no_of_moves +1
    print(f"{player_name} has {moves-no_of_moves} moves out of the {moves}\n")
print("GAME OVER")

if computer_score > player_score:
    print("Sorry!!! Computer won")
elif computer_score < player_score:
    print(f"Congratulations!!! {player_name} won")
else:
    print("Opps!!! It's Tie ")
print(f"{player_name} score is {player_score} and computer score is {computer_score}")
```
